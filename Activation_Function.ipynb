```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_plot(x, f, df, name):
    plt.figure()
    plt.figure(figsize=(12,6))
    plt.title(name, fontsize=20, fontweight='bold')
    plt.xlabel('z', fontsize=15)
    plt.ylabel('Activation function value', fontsize=15)
    sns.set_style("whitegrid")
    plt.plot(x, f, label="f (z)") # simple graph
    plt.plot(x, df, label="f '(z)") # differentiation graph
    plt.legend(loc=4, prop={'size': 15}, frameon=True,shadow=True, facecolor="white", edgecolor="black")
    plt.savefig(name + ".png")
    plt.show()
```

```python
z = np.arange(-10, 10, 0.01)
f = 1/(1 + np.exp(-z))
df = f*(1 - f)
make_plot(z, f, df, "Sigmoid")
```

```python
z = np.arange(-10, 10, 0.01)
f = np.tanh(z)
df = 1 - f*f
make_plot(z, f, df, "tanh")
```

```python
z = np.arange(-10, 10, 0.01)
f = z * (z > 0)
df = 1. * (z > 0)
make_plot(z, f, df, "ReLU")
```

```python
import math

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Tanh Activation Function
def tanh(x):
    return (2 / (1 + math.exp(-2 * x))) - 1

# ReLU Activation Function
def relu(x):
    return max(0, x)

# Leaky ReLU Activation Function
def leaky_relu(x, alpha=0.01):
    return max(alpha * x, x)

# Testing the activation functions with a sample value
x = -2.5

print("Sigmoid:", sigmoid(x))
print("Tanh:", tanh(x))         
print("ReLU:", relu(x))        
print("Leaky ReLU:", leaky_relu(x)) 
```
